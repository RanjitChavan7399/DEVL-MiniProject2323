```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(r"C:\Users\vaibhav\Downloads\swiggy1 (1).csv")
city_counts = df['City'].value_counts()
city_counts.plot(kind='bar')
plt.title('Number of Restaurants by City')
plt.xlabel('City')
plt.ylabel('Restaurant Count')
plt.show()
```

```python
%matplotlib inline

plt.title("Avg rating frequency")
plt.xlabel("Avg rating")
plt.ylabel("frequency")

city_count = df['Avg ratings'].value_counts()
city_count.plot(kind='bar')

plt.show()

```

```python
plt.hist(df['Price'], bins=10)
plt.title('Price Distribution')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
food_type_counts = df['City'].value_counts()
food_type_counts.plot(kind='pie', autopct='%1.1f%%')
plt.title('frequency of orders in each city')
plt.show()
```

```python
food_type_counts = df['Total ratings'].value_counts()
food_type_counts.plot(kind='pie', autopct='%1.1f%%')
plt.title('frequency of Total rating')
plt.show()
```

```python
top_cities = df['City'].value_counts().nlargest(10).index

# Filter the DataFrame for these top cities
filtered_df = df[df['City'].isin(top_cities)]

# Create a boxplot of Delivery Time grouped by City
filtered_df.boxplot(column='Delivery time', by='City')
plt.title('Delivery Time by City (Top 10 Cities)')
plt.xlabel('City')
plt.ylabel('Delivery Time (mins)')
plt.xticks(rotation=45)  # Rotate x labels for better visibility
plt.show()
```
